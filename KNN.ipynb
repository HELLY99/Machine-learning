```
from sklearn import datasets
import pandas as pd
```

```
# load iris dataset and extracting target feature from it
iris = datasets.load_iris()
data = iris.data
target = iris.target
```

```
iris.feature_names
```

```
# creating dataframe of iris dataset
dataset = pd.DataFrame(data , columns=iris.feature_names)
```

```
# Adding target variable into dataset
dataset['target'] = target
```

```
# converting numeric values into categorical feature
target_names = []
for i in range(0,len(dataset)):
  if dataset['target'][i] == 0:    
    target_names.append('setosa')         
  elif dataset['target'][i] == 1:
    target_names.append('versicolor')
  elif dataset['target'][i] == 2:
    target_names.append('virgnica')
```

```
dataset['labels'] = target_names
dataset.drop(['target'] , axis='columns' , inplace=True)
dataset
```

```
dataset.info()
```

```
import numpy as np
np.unique(dataset['labels'])
```

```
# Machine Learning model doesn't understand string values so we have to convert into numeric values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
le_target = encoder.fit_transform(dataset['labels'])
print(le_target)
```

```
#Removing labels columns from dataset
dataset.drop(['labels'] , axis = 'columns' , inplace=True)
```

```
dataset
```

```
le_target
```

```
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(dataset , le_target , test_size = 0.2 , random_state =3)
```

```
print(len(x_train))
print(len(x_test))
```

```
#import KNN model 
from sklearn.neighbors import KNeighborsClassifier
```

```
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(x_train , y_train)
```

```
prediction = classifier.predict(x_test)
```

```
# Accuracy score is performance matrics for this project
from sklearn.metrics import accuracy_score
print(accuracy_score(prediction , y_test))
```
